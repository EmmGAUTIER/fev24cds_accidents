```python
import sys
import pandas as pd
import re
import matplotlib.pyplot as plt

dfc = pd.read_csv("données_travail/caracteristiques_2.csv",
                 sep = '\t',
                 low_memory = False,
                 dtype = {"Num_Acc" : "object"})

#dfu = pd.read_csv("données_travail/usagers_1.csv",
#                 sep = '\t',
#                 low_memory = False,
#                 dtype = {"Num_Acc" : "object"})

```

```python
dfc.an = dfc.an + ((dfc.an<2000) *2000)
rep_annuelle = dfc.an.value_counts()
rep_annuelle = rep_annuelle.sort_index()

plt.plot(rep_annuelle)
plt.xlabel("Année")
plt.ylabel("Nombre")
plt.title("Évolution du nombre d'accidents corporels de la circulation en France")
plt.xticks (range(2006,2022, 2))
plt.show()
```
